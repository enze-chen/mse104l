```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.rcParams.update(mpl.rcParamsDefault)
```

```python
df = pd.read_csv('W_unused_Point.txt', skiprows=25, names=['Energy', 'Counts'], sep='\s+')

plt.rcParams.update({'figure.figsize':(9,5), 'font.size':18, 'lines.linewidth':2})
fig, ax = plt.subplots()
ax.plot(df['Energy'], df['Counts'], label='unused')
ax.annotate(text=r'C $K\alpha_1$', xy=(0.12, 250), fontsize=14)
ax.annotate(text=r'W $M\alpha_1$', xy=(2, 5510))
ax.annotate(text=r'W $L\alpha_1$', xy=(8.4, 600))
ax.annotate(text=r'W $L\beta_1$', xy=(8.8, 210))
ax.annotate(text=r'W $L\beta_2$', xy=(10.12, 100))

ax.set_xlim([0, 12])
ax.set_xlabel('Energy (keV)')
ax.set_ylim([9, 1e4])
ax.set_yscale('log')
ax.set_yticks([])
ax.set_ylabel('Intensity (a.u.)')
for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)
plt.show()
fig.savefig('W_unused_Point_replot.png', dpi=300, bbox_inches='tight')
```
